```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
import geodatasets
```

```python
# Filter for US states
gdf = gpd.read_file('ne_50m_admin_1_states_provinces.shp')
us_states = gdf[gdf['admin'] == 'United States of America']

# Plot the US states with boundaries and state names
fig, ax = plt.subplots(figsize=(15, 10))
us_states.boundary.plot(ax=ax, linewidth=1)
us_states.plot(ax=ax, column='name', legend=False, linewidth=0.8, edgecolor='black')

# Annotate the state names
for x, y, label in zip(us_states.geometry.centroid.x, us_states.geometry.centroid.y, us_states['name']):
    ax.text(x, y, label, fontsize=8, ha='center', color='darkred')

plt.title('US States with Boundaries and State Names')
plt.show()
```
